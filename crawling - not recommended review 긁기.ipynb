```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
import re
import numpy as np
```

```python
# 50개의 restaurant href는 그대로 사용이 가능
restaurant_href=pd.DataFrame.from_csv("restaurant_href_df.csv")
restaurant_href["restaurant_href"][49]
df=pd.DataFrame(columns=["not recommend url"])
```

```python
filter_name_df=pd.DataFrame(columns=["restaurant name"])
```

```python
for i in range(50):
    href=restaurant_href["restaurant_href"][i]
    restaurant_name= href.replace("/biz","")
    filter_name_df.loc[len(filter_name_df)] = [restaurant_name]
```

```python
filter_name_df.to_csv("filter_name_df.csv")
```

```python
#not recommended review의 pattern
# http://www.yelp.com/not_recommended_reviews/anchor-brewing-company-san-francisco
# http://www.yelp.com/not_recommended_reviews/anchor-brewing-company-san-francisco?not_recommended_start=10
for i in range(0,50):
    b =filter_name_df["restaurant name"][i]
    not_recommend= "http://www.yelp.com/not_recommended_reviews{item}".format(item=b)
    for k in range(0,10):
        k = k*10
        not_recommend_review= not_recommend+"?not_recommended_start={item}".format(item=k)
        df.loc[len(df)]=[not_recommend_review]
```
