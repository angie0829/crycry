```python
good_review= pd.DataFrame(columns=["restaurant name","url","review"])
good_review
```

```python
for index, row in restaurant_url_df.iterrows():
    url=row["url"]
    response = requests.get(url)
    try:
        dom = BeautifulSoup(response.content, "html.parser")
        review_elements = dom.select('div.review-content')
        restaurant_name_raw = dom.select_one('h1.biz-page-title.embossed-text-white.shortenough').text
        restaurant_name= restaurant_name_raw.replace('\n','')
    except:
        print(float('nan'))
            
    for review_element in review_elements:
        try:
            review= review_element.select_one('p').text
        except:
            print(float('nan'))
        good_review.loc[len(good_review)] = [restaurant_name,url,review]
        
```
